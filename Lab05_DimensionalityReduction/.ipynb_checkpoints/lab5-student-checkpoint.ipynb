#### 2.1.1 Scoring 
<div class="alert alert-info">
Use the scoring mechanism described in the lectures (only on the training data) to obtain a set of features (corresponding to a subset of the original ones). You should make sure you extract this subset from the test data too!
</div>

```python
X.shape, t.reshape(len(t)).shape
```

```python
def order_features_by_score(X, t):
    
    scores = np.zeros((X.shape[1], 2))
    
    for i in np.arange(X.shape[1]):
        class0 = X[:,i][(t==0).reshape(len(t))]
        class1 = X[:,i][(t==1).reshape(len(t))]
        
        score = np.abs(np.mean(class0) - np.mean(class1)) / (np.var(class0) + np.var(class1))
        
        scores[i,0] = int(i) # feature index
        scores[i,1] = score # score
    
    # Sort
    scores = np.array(sorted(scores, key=lambda x: x[1], reverse=True))
    
    ordered_features = [int(x) for x in scores[:,0]]
    
    return ordered_features, scores[:,1]
```

```python
ordered_features, scores = order_features_by_score(X, t)

print('Features:', ordered_features[:10])
print('Scores:', scores[:10])
```

#### 2.1.2 Clustering
<div class="alert alert-info">

You can use the $k$-means code you implemented in the previous notebook. Remember to pass the data the correct way around and only cluster the training data. Transforming the test data is a little more complex – your $k$-means should give you assignments of features to clusters. Use these alongside the test data to compute the cluster means for the test data (you do not need to re-assign the clusterings).

</div>

```python
from sklearn.metrics import pairwise_distances

def kmeans(X, k, maxiter, seed=None):
    c_history = list()

    # randomly choose k data points as initial centroids
    np.random.seed(seed)
    centroids = X[np.random.randint(0, X.shape[0]-1, size=k)][:]

    c_history.append(centroids)
    
    for itr in range(maxiter):
        # compute the distance matrix between each data point and the set of centroids
        distances_matrix = pairwise_distances(X, centroids)**2
        
        # assign each data point to the closest centroid
        cluster_assignment = np.argmin(distances_matrix, axis = 1)
        
        # Update centroids
        new_centroids = np.array([X[cluster_assignment == i].mean(axis = 0) for i in range(k)])
        
        # Verify stop condition
        if np.all(centroids == new_centroids):
            break
        centroids = new_centroids
        
        # Update centroids history
        c_history.append(centroids)
        
    return centroids, cluster_assignment, c_history
```

```python
n_features = 2
maxiter = 20

X_kmeans2, _, _ = kmeans(X.T, n_features, maxiter)

X_kmeans2 = X_kmeans2.T

# Test set
X_kmeans2_test, _, _ = kmeans(X_test.T, n_features, maxiter)
X_kmeans2_test = X_kmeans2_test.T
```

#### 2.1.3 Principal components
<div class="alert alert-info">
Implement a function for obtaining the $k$ principal components of your data.
Try to avoid using 'sklearn' libraries for this task!
</div>

```python
def pca_k(X, k):
    
    cov = np.cov(X)
    _, eigenvectors = np.linalg.eig(cov)
    return eigenvectors[:k]
```

## 3. Classification
<div class="alert alert-info">
Insofar, we have only considered how carrying out dimensionality reduction allows us to _visually_ distinguish the different classes of the dataset.<br><br>

Using any of the classification algorithms we developed in the previous lectures, assess how classification performance varies when carrying out any of the above feature selection techniques as a preprocessing step. </div>

```python
class NaiveBayesClassifier(object):
    def _estimate_class_prior(self, targets):
        '''
        Produces a dictionary of prior probabilities
        for the different class labels
        '''
        uniq, count = np.unique(targets, return_counts=True)
        self.prior = count / len(targets)
    
    def _estimate_params(self, x, y):
        '''
        Produces a dictionary of (MEAN, VAR) parameters
        for the gaussians of each class
        '''
        N = len(self.unique_classes)
        self.means = np.zeros(N, dtype=object)
        self.variances = np.zeros(N, dtype=object)
        
        for i, cls in enumerate(self.unique_classes):
            filtered = x[np.where(y == cls)]
            self.means[i] = np.mean(filtered, axis=0)
            var = np.var(filtered, axis=0)
            #var[var == 0] = 1e-10
            self.variances[i] = var

    def fit(self, X_train, y):
        '''
        Produces an array of tuples of the form (MEAN, STD)
        for each class labels. Those values will be used
        by the Naive classifier to make predictions
        '''
        # Store unique classes
        self.unique_classes = np.unique(y)
        
        # Estimate class prior
        self._estimate_class_prior(y)
        
        # Estimate class parameters
        self._estimate_params(X_train, y)
                
    def predict_proba(self, X):
        '''
        Return prediction probabilities for each unique class
        '''
        predictions = np.zeros((len(self.prior),len(X)))

        for i in range(len(self.prior)):
            gaussian_i = -0.5 * (np.sum(np.log(2. * np.pi * self.variances[i])) 
                                 + np.sum(((X - self.means[i])**2) / (self.variances[i]), axis=1))
            predictions[i] = gaussian_i * self.prior[i]
            
        # abs denominator because I don't want it to change the sign of my posterior
        predictions = np.array([p / abs(p.sum()) for p in predictions.T])
        
        return predictions
    
    def predict(self, X):
        '''
        Return prediction labels
        '''
        pred = self.predict_proba(X)
        return np.argmax(pred, axis=1)
```

```python
from sklearn.metrics import accuracy_score

techniques = ['Top 2 scoring features', 'Top 3 scoring features', 
    'K-Means with K=2', 'K-Means with K=3',
    'PCA with 2 components', 'PCA with 3 components']

data = [X[:,top2], X[:,top3],
    X_kmeans2, X_kmeans3, 
    X_pca2, X_pca3]

test = [X_test[:,top2], X_test[:, top3],
        X_kmeans2_test, X_kmeans3_test, 
        X_pca2_test, X_pca3_test]

print('Accuracies:')
for name, x, x_test in zip(techniques, data, test):
    clf = NaiveBayesClassifier()
    clf.fit(x, t)
    y_pred = clf.predict(x_test)
    print(' - {} -> {:.4f}'.format(name, accuracy_score(t_test, y_pred)))
```
