```python
# Import statements before gettings started
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt

import seaborn as sns

from pylab import rcParams
rcParams['figure.figsize'] = 10, 5
rcParams['font.size'] = 14
rcParams['axes.titlepad'] = 15 

import pandas as pd
import numpy as np

from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import mean_squared_error
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

import os
import itertools
import math
from time import time
```

```python
def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]

    fig, ax = plt.subplots(figsize=(8, 8))
    ax.imshow(cm, interpolation='nearest', cmap=cmap)
    ax.set_title(title, fontsize=15, weight='bold')
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    ax.set_ylabel('True label')
    ax.set_xlabel('Predicted label')
```

```python
def plot_pca_exp_var(x, dim):
    '''
    Utility function used to plot the explained variance
    variation by doing dimensionality reduction
    '''
    # Perform PCA and get explained variance
    pca = PCA(n_components=dim)
    x_pca = pca.fit_transform(x)
    exvar = (pca.explained_variance_ratio_)
    exvar.sort()
    exvar = exvar[::-1]
    
    # Compute CDF
    cdf = np.zeros_like(exvar)
    for i in range(len(exvar)):
        cdf[i] = exvar[0] if i == 0 else cdf[i-1]+exvar[i]
    cdf = np.insert(cdf, 0, 0)
    
    # Plot
    ax = plt.gca()

    # Find point in which we have 75% information
    f = np.where(cdf >= 0.75)[0][0]
    ax.axhline(0.75, color='orange', linestyle='-.',lw=2, alpha=0.5)
    ax.axvline(f, linestyle='-.', color='orange', lw=2, alpha=0.5)
    ax.text(f+5, 0.1, '%s features' % f)

    # Find point in which we have 99% information
    f = np.where(cdf >= 0.99)[0][0]
    ax.axhline(0.99, color='green', linestyle='-.',lw=2, alpha=0.5)
    ax.axvline(f, linestyle='-.', color='green', lw=2, alpha=0.5)
    ax.text(f+5, 0.1, '%s features' % f)

    plt.title('Explained Variance')
    plt.xlabel('Number of Components')
    plt.plot(np.arange(dim + 1), cdf, lw=2.5)
    plt.show()
```

## 3. Classification

### Naive Bayes Classifier

Before applying Naive Bayes classification to our datasets, let's define a class implementing this algorithm.

A thing which is worth mentioning is the fact that I have used log likelihood in order to ease the computation and to avoid underflow issues.

```python
class NaiveBayesClassifier(object):
    def _estimate_class_prior(self, targets):
        '''
        Produces a dictionary of prior probabilities
        for the different class labels
        '''
        uniq, count = np.unique(targets, return_counts=True)
        self.prior = count / len(targets)
    
    def _estimate_params(self, x, y):
        '''
        Produces a dictionary of (MEAN, VAR) parameters
        for the gaussians of each class
        '''
        N = len(self.unique_classes)
        self.means = np.zeros(N, dtype=object)
        self.variances = np.zeros(N, dtype=object)
        
        for i, cls in enumerate(self.unique_classes):
            filtered = x[np.where(y == cls)]
            self.means[i] = np.mean(filtered, axis=0)
            var = np.var(filtered, axis=0)
            var[var == 0] = 1e-10
            self.variances[i] = var

    def fit(self, X_train, y):
        '''
        Produces an array of tuples of the form (MEAN, STD)
        for each class labels. Those values will be used
        by the Naive classifier to make predictions
        '''
        # Store unique classes
        self.unique_classes = np.unique(y)
        
        # Estimate class prior
        self._estimate_class_prior(y)
        
        # Estimate class parameters
        self._estimate_params(X_train, y)
                
    def predict_proba(self, X):
        '''
        Return prediction probabilities for each unique class
        '''
        predictions = np.zeros((len(self.prior),len(X)))

        for i in range(len(self.prior)):
            gaussian_i = -0.5 * (np.sum(np.log(2. * np.pi * self.variances[i])) 
                                 + np.sum(((X - self.means[i])**2) / (self.variances[i]), axis=1))
            predictions[i] = gaussian_i * self.prior[i]
            
        # abs denominator because I don't want it to change the sign of my posterior
        predictions = np.array([p / abs(p.sum()) for p in predictions.T])
        
        return predictions
    
    def predict(self, X):
        '''
        Return prediction labels
        '''
        pred = self.predict_proba(X)
        return np.argmax(pred, axis=1)
```

### Strengths and Weaknesses
The Naive Bayes assumption is helpful when the dimensionality D of the input space is high (which is our case). Indeed, estimating D different univariate normal distributions is much easier than estimating one single multivariate normal distribution. In fact, in the first case we have to estimate 2D (D means and D variances) different parameters, while we need $D + D + \frac{D(D - 1)}{2}$ (D for the means and the rest for building the covariance matrix).

However, the main benefit of the Naive assumption is also its major downside. In fact, assuming that any dimension of the input is indipendent within classes may be a questionable statement and may lead to unprecise results.

### Data Preprocessing Suggestions

Below are reported some considerations about data preprocessing techniques which may be applied to both datasets.

#### Fashion MNIST

One thing which could be worth doing could be **binarizing the data**. Each pixel in our dataset has an integer value ranging from 0 to 255. Given a threshold value of 128 we could set to 1 those pixels having a value greater than 128 and to 0 the others. Let's see what happens if we apply this transformation to the first 8 elements of our dataset (their original version has already been plotted at the beginning of this notebook).

From the above images we see that it is not probably a great idea to transform our data into binary because some classes may look very similar between each other (Shirt and Coat or T-shirt/Top and Dress) and this may lead to ambiguity resulting in a decrease in our classifier performances.

Another thing which is worth trying is **dimensionality reduction**. Indeed, having 784 dimensions for distinguishing 10 classes may be quite a lot and PCA may help us. Let's see what the [explained variance](https://en.wikipedia.org/wiki/Explained_variation) is while applying PCA to our dataset.

```python
plot_pca_exp_var(X_train_fashionMNIST, 784)
```

The analysis of the cumulative explained variance suggests that, if we want to maintain the 75% of the original information, we should use 14 dimensions. However this amount of information may be too low, leading to a decrease in accuracy performances.

On the other hand, assuming that we can tolerate a 1% loss in precision (which is a reasonable assumption for this problem), it would be better to use 459 dimensions which will make us able to keep the 99% of the original information. 

Using dimensionality reduction would also improve the computational speed of our classifier as it will reduce the number of dimensions of the 40%.

Another preprocessing technique which may be applied to Fashion MNIST dataset could be **data normalization**. This could be easily achieved by simply dividing each of its input point by 255 (the maximum value they can achieve).

#### CIFAR10

With the CIFAR10 dataset trying to discritize our images (like we did with Fashion MNIST) makes no sense as the images are represented in a much more complex way with respect to those of the Fashion MNIST dataset.

Again, we probably have too many dimensions. Indeed we have input points with 3072 dimensions and we are trying to distinguish 10 different classes. Therefore, performing our **dimensionality reduction** analysis also on CIFAR10 could be certainly interesting.

```python
plot_pca_exp_var(X_train_cifar10, 3072)
```

The above plot shows us that with 21 dimensions we could have the 75% of the original information while, by using 658 dimension, we could lose just 1% of the original information. This is quite an interesting thing because it means that, by using ~20% of the original dimensions, we could still catch the 99% of the variation expressed in the original dataset. 

Another interesting thing which may be worth trying is [converting CIFAR10 data to **grayscale**](https://en.wikipedia.org/wiki/Grayscale#Converting_color_to_grayscale). Indeed, since CIFAR10 dataset contains colored images, converting those images to grayscale could make our model simpler while keeping more or less the same prediction accuracy.

Let's see how CIFAR10 images look like when we apply a grayscale transformation to it.

```python
def grayscale(X_cifar10):
    shape = (1024, X_cifar10.shape[0])
    result = np.zeros(shape)
    for i in range(shape[0]):
        #  0.21 R + 0.72 G + 0.07 B
        result[i] = X_cifar10[:, i]*0.21 + X_cifar10[:, i+shape[0]]*0.72 + X_cifar10[:, i+(2*shape[0])]*0.07 
    return result.T
```

```python
fig, axes = plt.subplots(2, 4, figsize=(15, 8))
for ax, label, image in zip(axes.flat, y_train_cifar10, grayscale(X_train_cifar10[:8])):
    im = image
    ax.imshow(im.reshape(32,32), cmap='gray')
    ax.set_title(cifar10_labels[label])
plt.show()
```

Figures belonging to different classes still seems to be very distinguishable. Therefore this transformation may lead us to good performance improvements.

We may also think of **binarizing grayscaled CIFAR10 images**. Let's see how those images would look like. 

```python
fig, axes = plt.subplots(2, 4, figsize=(15, 8))
for ax, label, image in zip(axes.flat, y_train_cifar10, grayscale(X_train_cifar10[:8])):
    im = image // 128
    ax.imshow(im.reshape(32,32), cmap='gray')
    ax.set_title(cifar10_labels[label])
plt.show()
```

From the above figure it seems very unlikely that we will be able to achieve any good results from using this technique. Indeed, figure are quite confused and they look a lot like random sets on black and white spots.

Similarly to what I said for Fashion MNIST, **data normalization** could be effective also in CIFAR10.

## 4. Bayesian Regression

### Bayesian Linear Regressor

Before applying Bayesian Regression to our datasets, let's define a class implementing this algorithm.

```python
class BayesRegressor(object):
    def __init__(self, order=1, n_samples=3):
        self.order = order
        self.N_SAMPLES = n_samples
        
    @staticmethod
    def get_X(x, n):
        '''
        Compute the matrix X featuring x raised to higher powers.
        Rescaling of the original inputs is recommended in order to avoid numerical instability.
        '''
        if n >= 4:
            x = x - np.mean(x)

        X_ = np.ones((len(x), 1))
        for i in range(1, n+1):
            X_ = np.column_stack((X_, np.power(x, i)))

        return X_
        
    @staticmethod
    def get_w_hat(x_, y, n):
        '''
        Compute optimal w given input
        '''
        return (np.linalg.inv(x_.T.dot(x_))).dot(x_.T).dot(np.array(y))

    @staticmethod
    def get_sigma2_hat(X, w_hat, t):
        sigma2_hat = (t - X.dot(w_hat)).T.dot(t - X.dot(w_hat)) / len(X)
        return sigma2_hat

    @staticmethod
    def get_cov_w(x, t, order):
        '''
        Compute covariance matrix
        '''
        X_ = BayesRegressor.get_X(x, order)

        w_hat = BayesRegressor.get_w_hat(X_, t, order)
        sigma2_hat = BayesRegressor.get_sigma2_hat(X_, w_hat, t)
        cov_w_hat = sigma2_hat * np.linalg.inv(X_.T.dot(X_))

        return cov_w_hat
        
    def fit(self, X, y):
        '''
        Estimate regression parameters given
        input examples
        '''
        # Compute optimal w and covariance matrix for a model of polynomial order self.order
        x_ = BayesRegressor.get_X(X, self.order)
        self.w_hat = BayesRegressor.get_w_hat(x_, y, self.order)
        self.cov_w = BayesRegressor.get_cov_w(X, y, self.order)
        
    def predict(self, X):
        '''
        Generate predictions for the given
        input based on the estimated parameters
        '''
        x_ = BayesRegressor.get_X(X, self.order)
        
        # Compute N_SAMPLES predictions and then average
        t_pred = np.zeros((self.N_SAMPLES, len(X)))
        for i in np.arange(self.N_SAMPLES):
            w_sample = np.random.multivariate_normal(self.w_hat, self.cov_w).T
            t_pred[i] = x_.dot(w_sample).reshape(-1)
            
        return np.mean(t_pred, axis=0)
```

### Fashion MNIST Regression

While applying the above implementation of a Bayesian regressor to the Fashion MNIST problem, the target labels will be considered as belonging to a continuos input space. The same will be done later for CIFAR10.

The built regressor will try to build a polynomial function of degree 1 (a linear function) to fit the input data. Predictions, instead, will be obtained by sampling from 3 normal distributions, built according to the parameters estimated during the training step.

Predictions are stored in a vector for later analysis.

```python
# Apply Bayes Regressor to the new training set
fashionMNIST_reg = BayesRegressor()
start = time()
fashionMNIST_reg.fit(X_train_fashionMNIST, y_train_fashionMNIST)
end = time()
print('Training completed in {:.02f} seconds'.format(end-start))
```

## 5. Convolutional Layers and Bayesian Classification

In this section I will experiment with Convolutional Neural Networks (CNN) combined with my implementation of the Naive Bayesian Classification algorithm.

CNNs are considered to be the state of the art in terms of image recognitions and, for this notebook, I will use an architecture made as follows:
 - Convolutional layer with 5x5 filter 
 - MaxPooling with a 2x2 kernel
 - Convolutional layer with 5x5 filter 
 - MaxPooling with a 2x2 kernel
 - Fully connected neuron layer with 500 neurons and ReLU activation function
 - Output layer with 10 neurons and softmax activation function
 
The first thing I will do will be to train the standalone CNN. After that I will detach the last two layers and apply the Naive Bayes Classifier to the output of the second convolutional layer (by convolutional layer I mean convolutional layer plus max pooling layer). Basically it will be like applying two trained convolutional filters to my input images as a sort of data preprocessing.

I will repeat the described experiment both for Fashion MNIST and CIFAR10. Each experiment will followed by a brief analysis of its results.

```python
# Inspired by: https://github.com/f00-/mnist-lenet-keras/blob/master/lenet.py
import keras
from keras.models import Sequential
from keras.layers.convolutional import Conv2D
from keras.layers.convolutional import MaxPooling2D
from keras.layers.core import Activation
from keras.layers.core import Flatten
from keras.layers.core import Dense
from keras.layers.core import Dropout

class LeNet(object):
    @staticmethod
    def build(width, height, depth, classes, 
              conv1_out=20, conv1_size=5, conv2_out=50, conv2_size=5):
        # initialize the model
        model = Sequential()

        # first set of CONV => RELU => POOL
        model.add(Conv2D(conv1_out, (conv1_size, conv1_size), padding='same',
            input_shape=(height, width, depth), data_format='channels_first', name='conv1'))
        model.add(Activation('relu'))
        model.add(MaxPooling2D(pool_size=(2, 2), name='maxpool1'))

        # second set of CONV => RELU => POOL
        model.add(Conv2D(conv2_out, (conv2_size, conv2_size), padding='same', data_format='channels_first', name='conv2'))
        model.add(Activation('relu'))
        model.add(MaxPooling2D(pool_size=(2, 2), name='maxpool2'))

        # set of FC => RELU layers
        model.add(Flatten(name='flatten'))
        
        model.add(Dense(500, activation='relu', name='dense'))
        model.add(Dropout(0.5))

        # softmax classifier
        model.add(Dense(classes, activation='softmax', name='output_layer'))

        return model
```

### Fashion MNIST

```python
from keras.models import Model
from keras.callbacks import EarlyStopping

def cnn_bayes(X_train, y_train, X_test, y_test, 
              w, h, depth, num_classes, batch_size, epochs, 
              conv1_out=20, conv1_size=5, conv2_out=50, conv2_size=5,
              verbose=True):
    # Reshape and normalize input to fit LeNet requirements
    X_train_conv = X_train.reshape(X_train.shape[0],
                                   w, h, depth)
    X_test_conv = X_test.reshape(X_test.shape[0],
                                   w, h, depth)

    # Convert class labels to binary vectors for LeNet
    y_train_vec = keras.utils.to_categorical(y_train, num_classes)
    y_test_vec = keras.utils.to_categorical(y_test, num_classes)
    
    # Build model
    model = LeNet.build(w, h, depth, num_classes, 
                        conv1_out, conv1_size, conv2_out, conv2_size)

    model.compile(loss=keras.losses.categorical_crossentropy,
                  optimizer='adam',
                  metrics=['accuracy'])
    if verbose:
        model.summary()

    # Define early stopping condition
    early_stop = EarlyStopping(monitor='val_acc', min_delta=1e-3, 
                               patience=1, verbose=verbose)

    model.fit(X_train_conv, y_train_vec,
              batch_size=batch_size,
              epochs=epochs,
              verbose=verbose,
              callbacks=[early_stop],
              validation_split=0.2)
    
    # Evaluate accuracy on the test set
    score = model.evaluate(X_test_conv, y_test_vec, verbose=False)
    print('[CNN]Test loss:', score[0])
    print('[CNN]Test accuracy:', score[1])
    
    # Cut the network after the convolutional layer
    layer_name = 'flatten'
    intermediate_layer_model = Model(inputs=model.input,
                                     outputs=model.get_layer(layer_name).output)

    # Generate intermediate predictions
    lenet_prediction_train = intermediate_layer_model.predict(X_train_conv)
    lenet_prediction_test = intermediate_layer_model.predict(X_test_conv)
    
    # Train Naive Bayes Classifer on intermediate predictions
    bc = NaiveBayesClassifier()
    bc.fit(lenet_prediction_train, y_train)
    
    # Generate final predictions
    return bc.predict(lenet_prediction_test)
```

```python
w, h = 28, 28
depth = 1
num_classes = 10

batch_size = 128
epochs = 5

# Run CNN + Naive Bayes
preds = cnn_bayes(X_train_fashionMNIST, y_train_fashionMNIST, 
                  X_test_fashionMNIST, y_test_fashionMNIST, 
                  w, h, depth, num_classes, batch_size, epochs, verbose=True)
```

```python
# Plot confusion matrix
cm = confusion_matrix(y_test_fashionMNIST, preds)
plot_confusion_matrix(cm, classes=fashionMNIST_labels, 
                    title='Confusion Matrix (LeNet + Naive Bayes Classifier)')
accuracy = (sum([cm[i,i] for i in range(len(cm))])) / len(preds)
print('Accuracy: %0.2f %%' % (accuracy*100))
```

Despite the fact that CNN alone produces good results (85% accuracy by running for just 3 epochs), combining the two algorithms leads to quite an improvement in terms of accuracy with respect to using just the Naive Bayes Classifier. Indeed, we moved from 57.53% to 80.60% accuracy.

Of course I could have given the CNN with more epochs and it would have probably reached much better results but this is not the goal of this notebook but this was omitted in the interest of computational time.
